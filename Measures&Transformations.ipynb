````python
import numpy as np
import pandas as pd
````

````python
main_df = pd.read_csv('full_df.csv',index_col = False)
main_df.info()
````

````python
Age_Group = []
for value in main_df['Patient Age']:
  if value < 30:
    Age_Group.append('<30')
  elif value < 30:
    Age_Group.append('30-44')
  elif value < 30:
    Age_Group.append('45-59')
  elif value < 75:
    Age_Group.append('60-74')
  else:
    Age_Group.append('> 75')
main_df['Age_Group'] = Age_Group

main_df.info()
````

````python
Primary_Disease = []
main_df['labels'] = [val[2] for val in main_df['labels']]
for value in main_df['labels']:
  if value == "N":
    Primary_Disease.append('Normal')
  elif value == 'D':
    Primary_Disease.append('Diabetes')
  elif value == 'G':
    Primary_Disease.append('Glaucoma')
  elif value == 'C':
    Primary_Disease.append('Cataract')
  elif value == 'A':
    Primary_Disease.append('AMD')
  elif value == 'H':
    Primary_Disease.append('Hypertension')
  elif value == 'M':
    Primary_Disease.append('Myopia')
  elif value == 'O':
    Primary_Disease.append('Other')
  else:
    Primary_Disease.append('Unkown')

main_df['Primary Disease'] = Primary_Disease #NOT_accurate :") but OK, it IS the

main_df.info()
````

````python
main_df['Disease Count'] = main_df[['N','D','G','C','A','H','M','O']].T.sum()
main_df.info()
````

````python
main_df['Disease_Count'][:4]
````

````python
main_df['Has Comorbidity'] = ["Yes" if value > 1 else "No" for value in main_df['Disease Count']]
main_df.info()
````

````python
main_df['Eye Side'] = ["Right Eye" if 'right' in value else "Left Eye" for value in main_df['filename']]
main_df.info()
````

````python
main_df['Eye Side'][:3]
````

````python
Total_Patients = len(set(main_df['ID']))
Total_Records = len(main_df.index)
Total_Normal_Cases = main_df['N'].sum()
Total_Diabetic_Cases = main_df['D'].sum()
Total_Glaucoma_Cases = main_df['G'].sum()
Total_Cataract_Cases = main_df['C'].sum()
Total_AMD_Cases = main_df['A'].sum()
Total_Hypertension_Cases = main_df['H'].sum()
Total_Myopia_Cases = main_df['M'].sum()
Total_Other_Cases = main_df['O'].sum()
````

````python
# Disease Prevalence % =
# DIVIDE(
# COUNTROWS('ODIR_Data'),
# CALCULATE(COUNTROWS('ODIR_Data'), ALL('ODIR_Data')),
# 0)  100
"""
Disease Prevalence % = DIVIDE( COUNTROWS('ODIR_Data'), CALCULATE(COUNTROWS('ODIR_Data'), ALL('ODIR_Data')), 0) *100

The provided DAX formula calculates the **percentage of total rows** in the `'ODIR_Data'` table, not a specific disease prevalence rate.

*   **COUNTROWS('ODIR_Data')**: Counts the number of rows in the current filter context (the numerator).
*   **CALCULATE(COUNTROWS('ODIR_Data'), ALL('ODIR_Data'))**: Counts all rows in the table, ignoring any filters (the denominator).
*   **Result**: This expression computes **% of Grand Total** for the current filter context.

To calculate **Disease Prevalence %**, you must restrict the numerator to only the rows where the disease status is positive (e.g., `1` or `"Yes"`). Assuming a column named `'ODIR_Data'[HasDisease]` indicates presence:

```dax
Disease Prevalence % =
DIVIDE(
    CALCULATE(COUNTROWS('ODIR_Data'), 'ODIR_Data'[HasDisease] = 1),
    CALCULATE(COUNTROWS('ODIR_Data'), ALL('ODIR_Data')),
    0
) * 100
```
"""
Normal_Rate = Total_Normal_Cases/ Total_Records * 100
Diabetes_Rate = Total_Diabetic_Cases/ Total_Records * 100
Glaucoma_Rate = Total_Glaucoma_Cases/ Total_Records * 100
Cataract_Rate = Total_Cataract_Cases/ Total_Records * 100
AMD_Rate = Total_AMD_Cases/ Total_Records * 100
Hypertension_Rate = Total_Hypertension_Cases/ Total_Records * 100
Myopia_Rate = Total_Myopia_Cases/ Total_Records * 100
````
